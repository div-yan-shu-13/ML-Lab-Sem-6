**Aim:** To implement Natural Language Processing (NLP) techniques such as Tokenization, Stop Word removal, Stemming, and Part-of-Speech (POS) tagging using the NLTK library in Python.

```python
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.probability import FreqDist

print("--- Step 1: Download NLTK Data ---")
# Downloading necessary NLTK datasets (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
print("NLTK datasets loaded successfully.\n")

# Sample text for our NLP pipeline
raw_text = """
Hello Mr. Smith, how are you doing today? The weather is absolutely great, and the city is awesome.
Artificial Intelligence and Machine Learning are transforming the technology landscape.
Python is an amazing programming language for building these intelligent systems.
"""
print("Original Text:")
print(raw_text.strip())
```

```python
print("--- Step 2: Tokenization ---")
# Sentence Tokenization
tokenized_sentences = sent_tokenize(raw_text)
print(f"Total Sentences: {len(tokenized_sentences)}")

# Word Tokenization
tokenized_words = word_tokenize(raw_text)
print(f"Total Words (including punctuation): {len(tokenized_words)}")
print("Sample Words:", tokenized_words[:10])
```

```python
print("--- Step 3 & 4: Stop Words Filtering & Stemming ---")
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()

filtered_words = []
stemmed_words = []

for word in tokenized_words:
    # Remove punctuation and stop words
    if word.casefold() not in stop_words and word.isalnum():
        filtered_words.append(word)
        # Apply Stemming to the valid words
        stemmed_words.append(stemmer.stem(word))

print(f"Words after filtering: {len(filtered_words)}")
print("Filtered Words Sample:", filtered_words[:8])
print("Stemmed Words Sample:", stemmed_words[:8])
```

```python
print("--- Step 5: Part-of-Speech (POS) Tagging ---")
# Tag the filtered words with their grammatical labels
pos_tags = nltk.pos_tag(filtered_words)

print("POS Tags Sample:")
for word, tag in pos_tags[:10]:
    print(f"{word:15} -> {tag}")
```

```python
print("--- Step 6: Frequency Distribution ---")
# Calculate word frequencies
fdist = FreqDist(filtered_words)

plt.figure(figsize=(10, 5))
fdist.plot(10, cumulative=False, title="Top 10 Most Common Words", color='black')
plt.show()
```
